```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.multiclass import OneVsRestClassifier

RANDOM_STATE = 42
TRAIN_SIZE = 0.75
```

```python
df = load_iris(as_frame=True)
x = df.data
y = df.target
```

```python
y.value_counts()
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, train_size=TRAIN_SIZE, random_state=RANDOM_STATE)
```

Применим логичестическую регрессию. У неё есть гиперпараметр multiclass, который принимает значения ovr или multinomial.  
ovr - one-versus-rest(один против всех) - сведение многоклассовой задачи к серии бинарных.  
multinomial - регение путем обобщения функции потерь логистической регрессии на многоклассовый случай

```python
model_ovr = OneVsRestClassifier(LogisticRegression(solver='lbfgs', random_state=RANDOM_STATE))
model_ovr.fit(xtrain, ytrain)
pred_ovr = model_ovr.predict(xtest)
accuracy_score(ytest, pred_ovr)
```

```python
model_multinomial = LogisticRegression(solver='lbfgs', random_state=RANDOM_STATE)
model_multinomial.fit(xtrain, ytrain)
pred_multinomial = model_multinomial.predict(xtest)
accuracy_score(ytest, pred_multinomial)
```

```python
print(classification_report(ytest, pred_ovr))
```

```python
print(classification_report(ytest, pred_multinomial))
```
